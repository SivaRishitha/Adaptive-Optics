```python
from sklearn.linear_model import Ridge
from sklearn.model_selection import LeaveOneOut
from sklearn.metrics import mean_squared_error

X = df[['InDia', 'Arm_angle', 'Fille_radius', 'arm length', 'OACD']]
Y = df[['4', '5', '6', '10', '11', '12', '13', '17']]
```

```python
from sklearn.linear_model import RidgeCV
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline

# try alphas over several orders of magnitude
alphas = np.logspace(-6, 1, 50)

pipe = Pipeline([
    ("scaler", StandardScaler()),
    ("ridge_cv", RidgeCV(alphas=alphas,
                         scoring="neg_mean_squared_error",
                         cv=LeaveOneOut()))
])

# For multi-output, RidgeCV handles it directly
pipe.fit(X, Y)
print("Best α:", pipe.named_steps["ridge_cv"].alpha_)
print("CV MSE:", -pipe.score(X, Y))
```

```python
model = Ridge(alpha=5.179474679231202)  # alpha is L2 regularization strength
model.fit(X, Y)

# Predict
Y_pred = model.predict(X)
error = mean_squared_error(Y, Y_pred)
print("Training Error:", error)
```

```python
from scipy.optimize import minimize

def inverse_predict(Y_target, model):
    def objective(x):
        x = np.array(x).reshape(1, -1)
        y_pred = model.predict(x)
        return np.mean((y_pred - Y_target) ** 2)

    # Initial guess (midpoint of each range)
    x0 = np.array([55.0, 40.0, 5.0, 8.0, 3.0])  # Midpoints

    # Explicitly defined bounds for the 5 features
    bounds = [
        (50.0, 60.0),      # Feature 1
        (37.5, 42.5),      # Feature 2
        (4.0, 6.0),   # Feature 3
        (6.0, 10.0),      # Feature 4
        (1.0, 5.0)   # Feature 5
    ]

    result = minimize(objective, x0, bounds=bounds)
    return result.x
```

```python
Y_target = np.array([100, 100, 100, 100, 100, 100, 100, 100])
X_estimated = inverse_predict(Y_target, model)
print("Estimated input features:", X_estimated)
```
